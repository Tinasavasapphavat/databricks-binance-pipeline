```python
import pandas as pd
import requests
from datetime import datetime,timedelta,timezone
import yaml
```

```python
URL = "https://api4.binance.com"
ENDPOINT = "/api/v3/aggTrades"

API = URL+ENDPOINT
print(API)
```

```python
start = int(datetime(2026,1,1,0,0,0,tzinfo=timezone.utc).timestamp()*1000)
end = int(datetime(2026,1,1,1,0,0,tzinfo=timezone.utc).timestamp()*1000)
```

```python
datetime.utcfromtimestamp(end/1000)
```

```python
SYMBOL = "BTCUSDT"

params = {
    "symbol" : SYMBOL,
    "startTime": start,
    "endTime": end ,
    "limit": 1000
}

res = requests.get(API,params=params)
data = res.json()
```

```python
pd.DataFrame(data)
```

```python
lst = []
end_ = end
while start < end:
    
    params = {
        "symbol" : SYMBOL,
        "startTime": start,
        "endTime": end ,
        "limit": 1000
        }

    res = requests.get(API,params=params)
    data = res.json()
    if len(data) == 1000:
        end = (start+end)//2
        # print(f"Reducing to: {datetime.utcfromtimestamp(start/1000).strftime("%Y-%m-%d %H:%M:%S")} - {datetime.utcfromtimestamp(end/1000).strftime("%Y-%m-%d %H:%M:%S")}")
    else:
        print(f"Fetching: {datetime.utcfromtimestamp(start/1000).strftime("%Y-%m-%d %H:%M:%S")} - {datetime.utcfromtimestamp(end/1000).strftime("%Y-%m-%d %H:%M:%S")}")
        df = pd.DataFrame(data)
        if len(df) != 0:
            lst.append(df)
        start = end+1
        end = (end_ + start)//2
lst
```

```python
df_raw = pd.concat(lst,axis =0).reset_index(drop=True)
df_raw
```

```python
df_raw.info()
```

```python
with open("../config/schema.yaml","r") as f:
    config = yaml.safe_load(f)
config 
```

```python
config['data']['columns'].items()
```

```python
def process_dataframe(df,config):
    for col , dt in config['data']['dtypes'].items():
        df[col] = df[col].astype(dt)
    
    df = df.rename(columns = config['data']['columns'])
    return df
```

```python
df_transform = process_dataframe(df_raw,config)
df_transform
```
